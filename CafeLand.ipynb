```python
#default function (send and get)BS4 *URL Require
def GetUrl(url):
    Uclient = Req(url)
    page_html = Uclient.read()
    Uclient.close()
    page_soup = Soup(page_html, "html.parser")
    return page_soup
```

```python
#Printing Profile by passing *html(URL) / GetUrl method
def Bs4Profile(Get):
#Name
    name = Get.find("h2",{"class":"b-sl-name"}).text
    print(name)
#Phone  
    for phone in Get.findAll('span',{'id':'detailTelSpan'}):
        phone = phone.text
        print(phone)
#Email
    email = Get.find('div',{'class':'col-md-4 col-sm-4'}).find("a").text
    print(email)

#Info
    for i in Get.findAll('div',{'class':'col-md-4 col-sm-4'}):
            info = i.find('p').text
    print(info)
#    
    Get.name= name
    Get.phone= phone
    Get.email= email
    Get.info= info
#FUnnysht
    if Get.name == Get.email:
        Get.name =Get.name+str(".")
    return {'name':[Get.name] , 'phone':[Get.phone] , 'Email':[Get.email] , 'Info':[Get.info]}
#Take Json Url    
Bs4Profile(GetUrl("https://nhadat.cafeland.vn/moi-gioi/namhung-at-bdsnamhungcom-0933334829-148234.html"))
```

```python
#select links in page /GetUrl method
def LinkGeneratorPlusProfile(Get):
        Storage = []
        for link in Get.findAll("div",{"class":"col-md-5"}):
            x = link.find("a").get("href")
            Storage.append(Bs4Profile(GetUrl(x)))
        return Storage
        
    
    #Export
        
        
        

LinkGeneratorPlusProfile(GetUrl("https://nhadat.cafeland.vn/moi-gioi/"))

    
```
